```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("calibrationtry2.csv", header = 0)
```

```python
df.columns.values
```

```python
df.shape
```

```python
df.iloc[0]
```

```python
X = df.drop(columns=['index', 'phbox', 'phprobe']).to_numpy()
y = df['phprobe'].to_numpy()
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
y_test
```

```python
reg.predict(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
np.average(np.subtract(reg.predict(X_test), y_test))
```
